```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.stem import SnowballStemmer
import matplotlib.pyplot as plt
import nltk
nltk.download('wordnet')

stemmer = PorterStemmer()
stemmer_espanol = SnowballStemmer('spanish')
lemmatizer = WordNetLemmatizer()

palabras = ["cantando", "saltando", "trabajando", "comiendo", "jugando", "durmiendo", "leyendo", "escribiendo", "bailando", 
            "nadando", "hablando", "cocinando", "dibujando", "riendo", "mirando", "cuidando", "pintando", "caminando", "esquivando", 
            "escuchando"]

palabrasIngles = ["singing", "jumping", "working", "eating", "playing", "sleeping", "reading", "writing", "dancing",
            "swimming", "talking", "cooking", "drawing", "laughing", "looking", "taking care", "painting", "walking", "dodging",
            "listening"]

print("Stemming: ")
for palabra in palabras:
    print(f"{palabra} → {stemmer_espanol.stem(palabra)}")

print("\nLematización: ")
for palabra in palabras:
    print(f"{palabra} → {lemmatizer.lemmatize(palabra, pos='v')}")

print("\n")
print("Stemming: ")
for palabra in palabrasIngles:
    print(f"{palabra} → {stemmer.stem(palabra)}")

print("\nLematización: ")
for palabra in palabrasIngles:
    print(f"{palabra} → {lemmatizer.lemmatize(palabra, pos='v')}")
```
